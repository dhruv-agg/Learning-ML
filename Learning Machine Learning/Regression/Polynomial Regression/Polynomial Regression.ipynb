```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read data
df = pd.read_csv("Position_Salaries.csv", sep = ',')
print("Data imported")
```

```python
df.head()
```

```python
X = df.iloc[:,1:2]
X.shape
```

```python
y = df.iloc[:,2]
y.shape
```

```python
plt.figure(figsize = (20,10))
plt.scatter(X,y,color = 'r')
plt.plot(X,y, label = 'Initial line')
plt.legend()
plt.ylabel('Salary')
plt.xlabel('Level')
plt.show()
```

```python
#Fit linear regr on X for comparison with Polynomial Regression
from sklearn.linear_model import LinearRegression
linregr = LinearRegression()
linregr.fit(X,y)
```

Since we have a small dataset, we won't split it in Test & Train

```python
#Visualising the Linear Regression
plt.figure(figsize = (20,10))
plt.scatter(X,y,color = 'r')
plt.plot(X, linregr.predict(X),color = 'black', label = 'Linear Regr')
plt.legend()
plt.ylabel('Salary')
plt.xlabel('Level')
plt.show()
```

X_poly
X_poly has a structure of 
- 1 (Constant)
- X^1
- X^2
- increases with degree

```python
#Proceeding to Polynomial Regression

from sklearn.preprocessing import PolynomialFeatures
poly_regr = PolynomialFeatures(degree = 2)
X_poly = poly_regr.fit_transform(X)
#transforms the X to a higher degree

linregr_poly = LinearRegression()
linregr_poly.fit(X_poly,y)
#Linear Regr fits on Polynomial X.

#Visualising the Linear Regression
plt.figure(figsize = (20,10))
plt.scatter(X,y,color = 'r')
plt.plot(X, linregr_poly.predict(X_poly),color = 'blue', label = 'Polynomial Regressor')
plt.legend()
plt.title("Polynomial Regressor Degree = 2")
plt.ylabel('Salary')
plt.xlabel('Level')
plt.show()
```
