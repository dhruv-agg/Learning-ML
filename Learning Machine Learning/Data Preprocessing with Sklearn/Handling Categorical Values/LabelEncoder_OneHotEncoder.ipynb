```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('data.xlsx')
print("Data imported")
```

```python
df.head(20)
```

```python
X = df.iloc[:,0]
Y = df.iloc[:,-1]
print('X',X)
print('Y',Y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
labelencoder = LabelEncoder() #create an object whenever you import a package
X = labelencoder.fit_transform(X)
X
```

**Cannot use LabelEncoder as we have multiple labels. Labels generated can be misunderstood as ranks for data. Thus use OneHotEncoder for more than 2 categories.**

```python
help(OneHotEncoder)
```

```python
#X = df.iloc[:,0]
#initialise X again before running encoder

X = X.reshape(-1, 1)
#do this to avoid ValueError when running OHE. 

onehotencoder = OneHotEncoder(categorical_features =[0], sparse = False)  #columns with categorical data
X = onehotencoder.fit_transform(X) #.to_array() Not needed as X is reshaped
#convert X into an array before feeding to encoder
print('X\n',X)
```

categorical_features : "all" or array of indices or mask
- |      Specify what features are treated as categorical.
- |      - 'all' (default): All features are treated as categorical.
- |      - array of indices: Array of categorical feature indices.

sparse = False
Makes the output more readable
